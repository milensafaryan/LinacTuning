```python
%matplotlib widget
aa = dfs[6].loc[:,'L:D21BF'].copy(deep=True)

threshold=300
aadiff = np.diff(aa)
disc_idx = np.where(abs(aadiff)>threshold)[0] 
if len(disc_idx)%2==1:
    disc_idx = np.append(disc_idx,len(aa)-1)
    
for i in range (1,len(disc_idx),2):
    for j in range(disc_idx[i-1],disc_idx[i]):
        if aadiff[disc_idx[i-1]]>0:
            aa[j+1] = aa[j+1]-360
        else:
            aa[j+1] = aa[j+1]+360
    
plt.plot(dfs[6].loc[:,'L:D21BF'])
plt.plot(aa)
```
